```python
import numpy as np
import scipy.io as spio

def convert_images_format(input_file_path, output_file_path):
    # 加载 .mat 文件
    data = spio.loadmat(input_file_path, squeeze_me=True)
    images = data['images']  # 假设图像数据存储在名为 'images' 的字段中
    angles = data['angles']  # 假设角度数据存储在名为 'angles' 的字段中

    # 检查原始形状
    print("Original shape:", images.shape)

    # 转换形状
    images_converted = np.transpose(images, (1, 2, 0))

    # 检查转换后的形状
    print("Converted shape:", images_converted.shape)

    # 保存到新的 .mat 文件
    spio.savemat(output_file_path, {'images': images_converted, 'angles': angles})

# 示例调用
input_file_path = 'data/output.mat'
output_file_path = 'data/updated_output.mat'
convert_images_format(input_file_path, output_file_path)
```

```python
import scipy.io as spio
import numpy as np
import tensorflow as tf
from scipy.ndimage import gaussian_filter

# Load the data
data_file_path = 'data/Diatom.mat'
data = spio.loadmat(data_file_path)
I_Raw = data['I_Raw']  # Assuming shape is (500, 500, 24)

# Define the Sobel edge detection function
def sobel_edges(image):
    sobel_x = tf.constant([[1, 0, -1], [2, 0, -2], [1, 0, -1]], dtype=tf.float32)
    sobel_y = tf.constant([[1, 2, 1], [0, 0, 0], [-1, -2, -1]], dtype=tf.float32)

    image = tf.reshape(image, [1, image.shape[0], image.shape[1], 1])  # Add batch and channel dimensions
    gx = tf.nn.conv2d(image, tf.reshape(sobel_x, [3, 3, 1, 1]), strides=[1, 1, 1, 1], padding='SAME')
    gy = tf.nn.conv2d(image, tf.reshape(sobel_y, [3, 3, 1, 1]), strides=[1, 1, 1, 1], padding='SAME')

    edges = tf.sqrt(tf.square(gx) + tf.square(gy))
    return tf.reshape(edges, [image.shape[1], image.shape[2]])

# Initialize an array to accumulate edge masks
accumulated_edges = np.zeros((700, 700), dtype=np.float32)

# Define the size of the border to ignore (e.g., 10 pixels)
border_size = 10

# Create a mask to ignore the borders
border_mask = np.ones((700, 700), dtype=np.float32)
border_mask[:border_size, :] = 0  # Top border
border_mask[-border_size:, :] = 0  # Bottom border
border_mask[:, :border_size] = 0  # Left border
border_mask[:, -border_size:] = 0  # Right border

# Use CPU for processing
with tf.device('/CPU:0'):  # 强制使用CPU
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        for i in range(I_Raw.shape[-1]):
            image = I_Raw[:, :, i]
            
            # Apply Gaussian smoothing to reduce noise
            smoothed_image = gaussian_filter(image, sigma=1.0)
            
            # Perform Sobel edge detection
            edges = sess.run(sobel_edges(smoothed_image))
            
            # Accumulate the edge mask
            accumulated_edges += edges

# Apply the border mask to ignore borders during normalization
masked_edges = accumulated_edges * border_mask

# Normalize the accumulated edge map to the range [0, 1] (excluding borders)
min_val = np.min(masked_edges[border_mask > 0])
max_val = np.max(masked_edges[border_mask > 0])
normalized_weight_map = (masked_edges - min_val) / (max_val - min_val)

# Set values below 0.1 to 0
normalized_weight_map[normalized_weight_map < 0.0001] = 0.0001

# Restore the ignored borders to 0
normalized_weight_map[border_mask == 0] = 0

# Save the normalized weight map to a .mat file
output_file_path = 'normalized_weight_map_diatom.mat'
spio.savemat(output_file_path, {'weight': normalized_weight_map})

```

```python
import scipy.io as spio
import matplotlib.pyplot as plt

# 加载 .mat 文件
file_path = 'normalized_weight_map_sigma2_diatom.mat'
mat_data = spio.loadmat(file_path)

# 遍历文件中的所有键，并输出键名和对应数据的形状
for key in mat_data:
    if not key.startswith('__'):  # 忽略系统键
        data = mat_data[key]
        print(f"Key: {key}, Shape: {data.shape}")
        
        # 如果数据是二维或三维的，尝试将其作为图像显示
        if len(data.shape) == 2:
            plt.imshow(data, cmap='gray')
            plt.title(f'Key: {key}')
            plt.colorbar()
            plt.show()
        elif len(data.shape) == 3:
            # 如果是三维的，显示其中第一层
            plt.imshow(data[:, :, 0], cmap='gray')
            plt.title(f'Key: {key} (First Layer)')
            plt.colorbar()
            plt.show()
```

```python
import os
import shutil

# 指定inference文件夹的路径
inference_dir = 'inference'  # 替换为你的inference文件夹路径
output_dir = 'test_image0'  # 替换为你想要保存图片的文件夹路径

# 如果输出文件夹不存在，则创建
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 遍历inference文件夹中的每个子文件夹
for subdir in os.listdir(inference_dir):
    subdir_path = os.path.join(inference_dir, subdir)
    
    if os.path.isdir(subdir_path):  # 检查是否是文件夹
        # 构建image_0.png的路径
        image_path = os.path.join(subdir_path, 'image_0.png')
        
        if os.path.exists(image_path):
            # 构建目标文件路径
            destination = os.path.join(output_dir, f"{subdir}_image_0.png")
            
            # 复制文件到目标文件夹
            shutil.copy(image_path, destination)
            print(f"已复制 {image_path} 到 {destination}")
        else:
            print(f"未找到 {image_path}")
```
